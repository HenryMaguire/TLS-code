# fig 2a Dynamics

```python
def prep_data_2a(alpha_over_eps):
    # alpha/epsilon = 0.1 -> filename = 'fig2a_alpha0p1eps'
    if alpha_over_eps <0.1:
        suff = "0"+str(int(alpha_over_eps*1000))
        if suff[-1] == '0':
            suff = suff[0:-1] 
    else:
        suff = str(int(alpha_over_eps*100))
        if suff[-1] == '0':
            suff = suff[0:-1]
    filename="ahsan-data/fig2a/fig2a_alpha0p{}eps.dat".format(suff)
    data = open(filename, 'r').read()
    times = []
    pops = []
    for row in data.split('\n'):
        cols = row.split('\t')
        times.append(float(cols[0]))
        pops.append(float(cols[1]))
    return times, pops
```

```python
alpha_ratios = [0.025, 0.05, 0.1, 0.15, 0.2, 0.25]
for i, alpha_over_epsilon in enumerate(alpha_ratios):
    times, pops = prep_data_2a(alpha_over_epsilon)
    if i == 0:
        plt.plot(times, pops, ls='dashed', color='k')
    else:
        plt.plot(times, pops)
plt.xlim(0,1)
plt.ylim(0,1)
plt.xlabel(r'Time (ns)')
plt.ylabel(r'$\expval{\sigma^{\dagger}\sigma}$')
plt.subplots_adjust(left=0.15, bottom=0.20, right=0.97, top=0.97, wspace=0.24, hspace=0.015)
plt.arrow(0.205, 0.157, 0.2, 0.3, linewidth=1.9, head_width=0.03, color='k')
plt.text(0.43, 0.39, r"Coupling strength")
plt.text(0.91, 0.9, "(a)", color='k', fontsize=21, fontweight='bold')
plt.savefig("thesis-figures/fig_2a.pdf")

```

# fig 2b Decay rates

```python
def prep_data_2b():
    # alpha/epsilon = 0.1 -> filename = 'fig2a_alpha0p1eps'
    filename="ahsan-data/fig2b/fig2b_60states.dat"
    data = open(filename, 'r').read()
    couplings = []
    rates = []
    for row in data.split('\n'):
        cols = row.split('\t')
        couplings.append(float(cols[0]))
        real_part = float(cols[1][0:16])
        rates.append(real_part)
    return couplings, rates

couplings, rates = prep_data_2b()

plt.plot(couplings, rates, color=colors[1])
plt.axhline(1, ls='dashed', color=colors[0])
plt.xlim(0,0.3)
plt.ylabel(r"Total decay rate, $\Gamma_0^{-1}$")
plt.xlabel(r"Coupling strength, $\alpha\epsilon^{-1}$")
#my_rates = load_obj('thesis-data/rates-1')
#plt.plot(my_rates[0], my_rates[1])

#my_rates = load_obj('thesis-data/rates-2')
#plt.plot(my_rates[0], my_rates[1])
plt.subplots_adjust(left=0.15, bottom=0.20, right=0.96, top=0.97, wspace=0.24, hspace=0.015)
plt.text(0.275, 0.9, "(b)", color='k', fontsize=21, fontweight='bold')
plt.savefig("thesis-figures/fig_2b.pdf")
```

# fig 3a Temperature dependence 1

```python
def prep_data_3a(T, N):
    filename="ahsan-data/fig3aT{}convergence/fig3aT{}_{}states.dat".format(int(T), int(T), int(N))
    data = open(filename, 'r').read()
    couplings = []
    pops = []
    for row in data.split('\n'):
        cols = row.split('\t')
        couplings.append(float(cols[0]))
        pops.append(float(cols[1]))
    return couplings, pops
```

```python
temps = [6000., 12000, 60000]
colors_3a = [colors[1], colors[9], colors[0]]
for T, color in zip(temps, colors_3a):
    ahsan_coups, ahsan_pops = prep_data_3a(T, 100)
    plt.axhline(ahsan_pops[0], c=color, linestyle='dashed', linewidth=1.6)
    plt.plot(ahsan_coups, ahsan_pops, ls='solid', c=color)
    plt.ylim(0,1)
    plt.xlim(0, ahsan_coups[-1])
    plt.subplots_adjust(left=0.16, bottom=0.20, right=0.97, top=0.97, wspace=0.24, hspace=0.015)
plt.xlabel(r'Coupling Strength, $\alpha\epsilon^{-1}$')
plt.ylabel(r'$\expval{\sigma^{\dagger}\sigma}_{ss}$')
plt.text(0.03, 0.91, "(a)", color='k', fontsize=21, fontweight='bold')
plt.savefig("thesis-figures/fig_3a.pdf")
```

# Fig 3b Temperature dependence 2

```python
def prep_data_3b():
    filename="ahsan-data/fig3b/fig3b.dat"
    data = open(filename, 'r').read()
    couplings = []
    pops = []
    for row in data.split('\n'):
        cols = row.split('\t')
        couplings.append(int(cols[0]))
        real_part = float(cols[1][0:16])
        if real_part > 1:
            real_part*=1e-6
        pops.append(real_part)
    return couplings, pops

couplings, pops = prep_data_3b()
```
